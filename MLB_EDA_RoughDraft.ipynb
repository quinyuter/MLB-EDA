#### Analysis of Variance

Next, I want to perform an Analysis of Variance on Wins to see if Year truly influences wins. While I don't have experience doing One-Way ANOVAs in Python, I have experience performing them in R, which is why I am especially curious to see how it works in this language, and more importantly, what it tells us about the data. 

Source: https://www.geeksforgeeks.org/how-to-perform-a-one-way-anova-in-python/

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Performing One-Way ANOVA (Year influences Wins)
# Year needs to be treated as a categorical variable
model = ols('Wins ~ C(Year)', data=bsbl_df).fit()

# Perform the ANOVA using statsmodels
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```

Looking at the p-value above, we see that it is 0.931618. This is much above the usual alpha value of 0.05, so we conclude that Year does not influence wins (i.e. there is no singificant difference of Wins between Years). This output makes sense to me in relation to the plots above. There does not seem to be many trends in terms of Wins over years.

<br>
Now that we know there aren't trends in terms of wins over years, I want to take things in a different direction. I would like to see which statistics are significant in predicting Wins. I then want to use these statistics to see if they trend in any certain direction. To do so, I want to run multiple linear regression on the offensive and defensive stats and compare them with their correlation matrices to see if we can come to any conclusion regarding offensive and defensive stats in relation to wins.

#### Multiple Linear Regression

```python
offense_measures = ['Runs_Per_Game','Plate_Appearances', 'At_Bats', 'Runs', 'Hits', 'Doubles', 'Triples', 'HR', 'RBI', 'SB', 'Caught_Stealing', 'Walks', 'Strikeouts', 'Batting_Average', 'On_Base_Percentage', 'Slugging_Percentage', 'On_Base_Plus_Slugging']
defense_measures = ['Errors', 'Double_Plays', 'Fielding_Percentage']
```

##### Offensive Stats

```python
# Define the formula for multiple linear regression
formula = 'Wins ~ ' + ' + '.join(offense_measures)

# Fit the regression model
off_model = ols(formula, data=bsbl_df).fit()

# Print the summary of the regression model
print(off_model.summary())
```

This gives us a lot of really cool information. Based on the p-values given, and using the alpha value of 0.05, we see that the offensive stats that affect wins are Doubles, Triples, Homeruns, and Slugging Percentage. I find this incredibly interesting because the only variables that are statistically significant are ones that relate to power hitting (because regular hits are not included, nor are stats like Batting Average and On Base Percentage). This is surprising to me because I would think that more runs equals more wins, but this test shows us that this assumption may not be true. I am also shocked to see that OPS is not influential, because that is a pretty all encompassing statistic regarding hitting. I am curious to see if the correlation matrix gives similar answers.

We also get a note that there is strong multicollinearity between the variables, which we will explore in the correlation matrix below. I understand that this does affect the model, which is why I know this method isn't perfect. It was just another way to see how the data are related to each other.

Note: The Adjusted R-Squared is 0.583, which isn't bad but isn't great. The purpose of this isn't to fit a prediction model, but instead to just see which variables are significant. If I were to be doing more models and eventual prediction, an Adjusted R-Squared value of 0.583 is not a value I feel is high enough.

```python
# Adding Wins
off_wins = offense_measures + ['Wins']
corr_off = bsbl_df[off_wins].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_off, annot=True, cmap='RdBu', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix for Offensive Baseball Statistics, Along With Wins')
```

Looking at just the Wins row, we see something very interesting. Doubles and Homeruns have just a moderate positive correlation with wins. But, triples has a nearly 0 correlation, meaning wins and triples aren't really correlated at all. Slugging has a 0.65 correlation, which is almost strong, but not quite. Meaning, Runs per Game, Runs, RBI, OBP, and OPS are much more correlated. These are the variables that I initially thought would be statistically significant in the model above. This further proves the idea that correlation is not causation. I am just surprised at how different these responses really are. To get back to trend analysis, I am going to use the statistically significant variables from above to present a line graph, in relation to wins. I am not choosing the correlated variables because they will present similarly as wins.

Looking at the other variables, we see that there are some very strong correlations. OBP, Slugging, and OPS are all very strongly positively correlated to Runs Per Game, Runs, and RBI. Batting Average is strongly positively correlated with Hits. All of these make sense because if these measures are high, more runs will score. Runs Per Game is 100% positively correlated with (1.00 correlation value) both Runs and RBIs, which makes sense, since Runs refers to the number of runs and RBIs are runs batting in (runs scoring). Then we get Strikeouts being moderately strongly negatively correlated with Hits and Batting Average. This makes sense also, because if one strikes out, they don't get a hit and their batting average decreases.

As stated above, the MLR output gave us a note saying there is multicollinearity, which is proven in the correlation matrix again. This makes sense, because Doubles, Triples, and Homeruns are all part of Slugging Percentage.

```python
## Running a linear regression to see if salary is important in predicting wins
import statsmodels.formula.api as smf

model_salary = smf.ols(formula='Wins ~ Salary', data=bsbl_df).fit()

# Print the model summary
print(model_salary.summary())
```

We aren't using this as our only model to predict wins, I was just more curious about the p-value. And, we now know that Salary is a statistically significant predictor. We also get the note that there is strong multicollinearity, which I find a bit confusing. When I think of multicollinearity, I think of statistics like On Base Percentage and On Base Plus Slugging Percentage, since On Base Percentage directly creates On Base Plus Slugging.
